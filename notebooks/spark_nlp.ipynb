```python
# this one works
from pyspark.sql import functions as F
from pyspark.sql.types import DateType
from pyspark.ml import Pipeline
from sparknlp.base import DocumentAssembler
from sparknlp.annotator import MultiDateMatcher
import sparknlp
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("YourAppName") \
    .config("spark.driver.memory", "8g") \
    .config("spark.executor.memory", "8g") \
    .getOrCreate()

# Start Spark NLP session
#spark = sparknlp.start()


# Create a DataFrame for the text list
df = spark.createDataFrame([(text,) for text in text_list], ["Text"])

# Spark NLP pipeline components
document_assembler = DocumentAssembler() \
    .setInputCol("Text") \
    .setOutputCol("document")

# Using MultiDateMatcher for multiple date extraction
multi_date_matcher = MultiDateMatcher() \
    .setInputCols("document") \
    .setOutputCol("dates") \
    .setOutputFormat("dd/MM/yyyy").setSourceLanguage("fr")

pipeline = Pipeline(stages=[document_assembler, multi_date_matcher])

# Fit and transform the pipeline
model = pipeline.fit(df)
result = model.transform(df)

# Convert 'dates.result' to DateType
# Create a new column where we convert the dates to actual DateType (in case it's in string format)
result = result.withColumn(
    "dates_converted",
    F.expr("transform(dates.result, x -> to_date(x, 'dd/MM/yyyy'))")
)

# Extract the most recent date from the list of dates
# We use 'greatest' to find the latest date in the list
result = result.withColumn(
    "most_recent_date",
    F.expr("greatest(" + ",".join([f"dates_converted[{i}]" for i in range(10)]) + ")")  # assuming max 10 dates
)

# Format the most recent date back to 'dd/MM/yyyy' string format
result = result.withColumn(
    "most_recent_date_formatted",
    F.date_format("most_recent_date", "dd/MM/yyyy")
)

spark.conf.set("spark.sql.debug.maxToStringFields", 500)
# Show the results
result.select("Text", "most_recent_date_formatted").show()

```
